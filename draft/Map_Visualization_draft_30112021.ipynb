```python
import geopandas as gpd
#import pandas as pd
#import matplotlib.pyplot as plt
import numpy as np
```

```python
land = gpd.read_file("vg2500_krs.shp")
```

```python
land = land[223:319] # We only need the map from Ingolstadt to Oberallgäu
```

```python
# delete these columns as they are useless to us
land = land.drop('RS_ALT', axis = 1)
land = land.drop('USE', axis = 1)
land = land.drop('RS', axis = 1)
land = land.drop('SHAPE_LENG', axis = 1)
land = land.drop('SHAPE_AREA', axis = 1)

# uncomment this line to see what the GeoDataFrame looks like
land
```

```python
# add pandemic statistics, only for testing (because we do not have the statistic yet)
number = []
for i in range(96):
    r = np.random.randint(5000,20000)
    number.append(r)
```

```python
# add the pandemic statistics
land['number'] = number
land
```

```python
# visualize the map
land_v = land.explore(
    column='number',
    cmap='Reds',
    scheme='NaturalBreaks',
    vmin=0,
    highlight = True,
    legend_kwds={
        'caption': 'pandemic',
        'scale': True,
        'colorbar' : False
    },
    style_kwds={
        'color': 'white',
        'weight': 1,
        'fillOpacity': 0.7
    },
    highlight_kwds={
        'fillColor': 'white',
        'fillOpacity': 0.6
    }
)

land_v
```
